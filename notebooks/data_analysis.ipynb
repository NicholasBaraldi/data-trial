## Starting the analysis

I'll start by ranking the top 10 stores that have at least 10 reviews.

```python
all_time_rank = pd.read_sql(
    """
    select cpg.name, count(crg.reviews) as total_reviews, round(avg(crg.review_rating)::numeric, 2) as avg_review
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    where crg.reviews >= 10
    group by cpg.name
    order by avg_review desc, total_reviews desc
    limit 10
    """,
    con=engine
)
all_time_rank
```

Okay, so these are the best ones already? Not exactly!\
As you can see below, the total data period is quite large, so it's possible that many stores that received good ratings in the past are bad today, or vice versa.

```python
full_period = pd.read_sql(
    """select min(review_datetime_utc), max(review_datetime_utc)
    from staging.customer_reviews_google
    """,
    con=engine
)
full_period
```

So to avoid this problem, I will stick to the year 2023.\
I chose 2023 because it is the most recent closed year, but it is possible to do this same analysis for any period.

Now a ranking of the 10 best stores in the year 2023 that have at least 10 reviews.

```python
overall_recent_rank = pd.read_sql(
    """
    select cpg.name, count(crg.reviews) as total_reviews, round(avg(crg.review_rating)::numeric, 2) as avg_review
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    where review_datetime_utc between '2023-01-01' and '2023-12-31'
    and crg.reviews >= 10
    group by cpg.name
    order by avg_review desc, total_reviews desc
    limit 10
    """,
    con=engine
)
overall_recent_rank
```

It's interesting to note that the top 10 has changed almost completely, the only company that remained relevant was "Pure Moving Company Miami Movers Local & Long distance", even maintaining the first position.

Now I will rank the best stores by city in 2023 that have at least 10 reviews.

```python
ranking_by_city = pd.read_sql(
    """
    with q as (
	select row_number() over(partition by cpg.city order by round(avg(crg.review_rating)::numeric, 2) desc, count(crg.reviews) desc) as rank, cpg.name, cpg.city, count(crg.reviews) as total_reviews, round(avg(crg.review_rating)::numeric, 2) as avg_review
	from staging.company_profiles_google_maps cpg
	join staging.customer_reviews_google crg on cpg.name = crg.name
	where review_datetime_utc between '2023-01-01' and '2023-12-31'
	group by cpg.name, cpg.city
	having count(crg.reviews) >= 10
	)
	select rank, city, name, total_reviews, avg_review from q
	where rank <= 10
	order by city, rank asc
    """,
    con=engine,
    index_col="rank"
)

ranking_by_city
```

You may notice that Dallas doesn't have the full ranking, there probably aren't 10 stores with 10 reviews in 2023.

I noticed that many companies have a rating of 5, so next I want to see what the rating distribution is.

```python
notes = pd.read_sql(
    """select count(*) as total_reviews, crg.review_rating
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by crg.review_rating
    order by crg.review_rating desc
    """,
    con=engine
)
notes
```

Based on the result we can conclude that almost no one writes an average review, most people only feel the need to express themselves when it is one of the extremes, 1 or 5 with a very strong emphasis on 5, which is chosen more than 90% of the time.

My next step is to see if a company having an "About" section influences the rating and number of reviews the company receives.

I'll group them by store and take the average number of reviews and ratings.

```python
has_about = pd.read_sql(
    """
    with q as(
        select (cpg.about::jsonb != '{}'::jsonb) as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
        from staging.company_profiles_google_maps cpg
        join staging.customer_reviews_google crg on cpg.name = crg.name
        group by has_about, cpg.name
    )
    select has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="has_about"
)
has_about
```

```python
sns.set_theme(style="whitegrid")
sns.catplot(data=has_about, kind="bar", x="has_about", y="avg_rating", hue="has_about", palette="dark", alpha=.6, height=6)
sns.catplot(data=has_about, kind="bar", x="has_about", y="avg_reviews", hue="has_about", palette="dark", alpha=.6, height=6)
```

We can see that the store having an "About" does not influence the rating that much, but there is a considerable increase in the number of reviews, probably indicating that it has more customers, so it is a good idea to have an "About".

Okay, so an "About" tag helped a little in customer acquisition, but is there a tag that helps more than others? Let's see what we have.

```python
about_tags = pd.read_sql(
    """
    select distinct(json_object_keys(about)) as about_tags
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    """,
    con=engine
)
about_tags
```

Okay, now let's see how much each tag influences.

```python
accessibility_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Accessibility') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Accessibility' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

service_options_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Service options') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Service options' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

amenities_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Amenities') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Amenities' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

payments_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Payments') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Payments' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

from_the_business_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'From the business') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'From the business' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

other_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Other') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Other' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

planning_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Planning') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Planning' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

crowd_df = pd.read_sql(
    """
    with q as(
    select (cpg.about::jsonb ? 'Crowd') as has_about, count(crg.*) as reviews, avg(crg.review_rating) as rating, cpg.name
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    group by has_about, cpg.name
    )
    select 'Crowd' as tag, has_about, round(avg(reviews)) as avg_reviews, round(avg(rating)::numeric, 2) avg_rating
    from q
    group by has_about
    """,
    con=engine,
    index_col="tag"
)

data_frames = [accessibility_df, service_options_df, amenities_df, payments_df, from_the_business_df, other_df, planning_df, crowd_df]
abouts_relevance = pd.concat(data_frames)
abouts_relevance
```

Based on this table, we can highlight a few things.
- "Service options" and "Crowd" significantly increase the number of reviews a store receives, so these are probably things that customers look for when hiring a company's services.
- "Amenities" and "Crowd" were the ones that most increased the average rating of companies, even though it wasn't much.
- Despite the significant change in the number of reviews, none of them significantly changed the average rating of companies.

We can therefore conclude that tags help attract customers but do not necessarily increase the rating that the customer will give.
In addition, "Crowd" is probably the most relevant tag for a company.

Finally, now that we have found what is important, let's take a look at "Pure Moving Company Miami Movers Local & Long distance" which is in first place.

```python
pd.set_option('display.max_colwidth', None)
best_company = pd.read_sql(
    """
    select cpg.name, cpg.city, cpg.phone, cpg.site, cpg.state, cpg.about
    from staging.company_profiles_google_maps cpg
    join staging.customer_reviews_google crg on cpg.name = crg.name
    where cpg.name like 'Pure Moving Company Miami Movers Local & Long distance'
    limit 1
    """,
    con=engine
)
best_company
```

As we can see, it is a verified company, has both a telephone number and a website, and the tags I highlighted previously, so it is a safe company, with easy access and that will please the end consumer, and its position in the ranking demonstrates this!
